```python
import pandas as pd
car_sales_missing = pd.read_csv('.\documents\car_sales_missing_data.csv')
#car_sales_missing['Odometer'] = car_sales_missing['Odometer'].fillna(car_sales_missing['Odometer'].mean
def dollar(value):
    if value == int:
        return int(value.replace(value))/100
    return value
car_sales_missing['Price'] = car_sales_missing['Price'].replace('[\$]', '', regex=True).apply(dollar)
car_sales_missing


```

```python
car_sales_missing['Wheel'] = 4
car_sales_missing['Country'] = pd.Series(['Japan', 'Japan', 'Japan', 'Germany', 'USA'])
car_sales_missing['Year'] = [1983,2012,2022,2023,1998,2007,2009,2016,1900,2000] ## list has to match
car_sales_missing['Size'] = car_sales_missing['Odometer'] * car_sales_missing['Doors'] * car_sales_missing['Wheel']
car_sales_missing['Size'].fillna('Missing', inplace=True)
car_sales_missing
```
