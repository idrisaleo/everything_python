```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
car_sales = pd.read_csv('./documents/car_sales.csv')
car_sales['Price'] = car_sales['Price'].str.replace('[\$\, \.]', '')
car_sales['Price'] = car_sales['Price'].str[:-2]
#car_sales
```

```python
car_sales['Sales Date'] = pd.date_range('29/06/2022', periods=len(car_sales))
car_sales
```

```python
car_sales['Total Sales'] = car_sales['Price'].astype(int).cumsum()
car_sales
```

```python
car_sales['Price'] = car_sales['Price'].astype(int)
car_sales.groupby('Make')['Price'].plot(kind='bar', legend=True) ## to line plot

```

```python
group_make_price = car_sales.groupby('Make')['Price']
```

```python
car_sales.plot(x='Make', y='Price', kind='scatter') ##scatter on pandas
```

```python
car_sales.pivot(columns='Make',
               index='Sales Date',
               values='Price').plot(kind='bar', figsize=(15,5))
```

```python
heart_disease = pd.read_csv('./documents/heart_disease.csv')
over_50 = heart_disease[heart_disease['age'] > 50]
```

```python
## pyplot method

over_50.plot(kind='scatter', x='age', y='chol', 
             c='target', figsize=(10,6))
```

```python
## OOP method
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)
```
