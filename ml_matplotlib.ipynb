## Matplotlib workflow


```python
# 0. Import matplotlib
import matplotlib.pyplot as plt

# 1. Prepare your data
x = [1,2,3,4,5]
y = [40,5,66,7,12]

## 2. Set up your plot

fig, ax = plt.subplots(figsize=(5,5)) ## width, height

## 3. plot data

ax.plot(x,y) ## with x and y repping the data

## 4. customize data

ax.set(title='example matplotlib',
      xlabel='x-axis label',
      ylabel='y-axis label')

## 5. Save plot
fig.figsave = ('documents/example_matplotlib.png')

```

```python
car_sales = pd.read_csv('./documents/car_sales.csv')
car_sales['kk'] = car_sales['Price'].str.replace('[\$\,\.]', '')
car_sales['Price'] = car_sales['kk'].str[:-2]
car_sales.drop(columns='kk')
```

```python
def inset(input):
    if input == int:
        return input
    if 'K' in input:
        if len(input) > 1:
            return input(input.replace('K', '')) * 1000
        return 1000
    if 'M' in input:
        if len(input) > 1:
            return input(input.replace('M', '')) * 1000000
        return 1000000
    if 'B' in input:
        if len(input) > 1:
            return input(input.replace('B', '')) * 1000000000
        return 1000000000
```
