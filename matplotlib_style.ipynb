```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd 
plt.style.available
plt.style.use('dark_background')
```

```python
np.random.seed(0)
x = np.random.rand(10,4)
df = pd.DataFrame(x, columns=list('ABCD'))
ax = df.plot(kind='bar')
ax.set(title='Random Bar Graph',
      xlabel='x-axis',
      ylabel='y-axis')
ax.legend().set_visible(True)
```
