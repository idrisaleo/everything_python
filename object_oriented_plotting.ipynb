```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
heart_disease = pd.read_csv('./documents/heart_disease.csv')
over_50 = heart_disease[heart_disease['age'] > 50]
```

```python
fig, ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter', x='age', y='chol', c='target', ax=ax)
```

```python
##Set Plot Style
plt.style.use('seaborn-whitegrid')

## OOP plotting method

fig, ax = plt.subplots(figsize=(10,6))
scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'],
                    cmap='winter') ## changes colour scheme

## setting up the axes

ax.set(title='Predicting heart disease based on cholesterol level and age',
      xlabel='age',
      ylabel='cholesterol level')

## adding legend

ax.legend(*scatter.legend_elements(), title='Target')

## adding horizontal line to figure out the average chol level....

ax.axhline(y=over_50['chol'].mean(), linestyle='--');
```

```python
fig, (ax0, ax1) = plt.subplots(ncols=1, nrows=2, figsize=(10,10), sharex=False) 
## with sharex, you can remove the xaxis, that should be after after sha

## plotting for first axis

scatter1 = ax0.scatter(x=over_50['age'],
           y=over_50['chol'],
           c=over_50['target'],
                      cmap='winter')

ax0.set(title='Age and Cholesterol',
       xlabel='Age',
       ylabel='Cholesterol')

##adding limit and erasing lines
ax0.set_xlim([50,80])
ax0.set_ylim([20,650])


ax0.legend(*scatter.legend_elements(), title='Target')

ax0.axhline(y=over_50['chol'].mean(), linestyle='-.')

## plotting for second axis

scatter2 = ax1.scatter(x=over_50['age'],
                       y=over_50['thalach'],
                       c=over_50['target'],
                      cmap='winter')

ax1.set(title='Age and Thalach',
       xlabel='Age',
       ylabel='Thalach')

ax1.set_xlim([50,80])
ax1.set_ylim([80,200])

ax1.legend(*scatter2.legend_elements(), title='Target')

ax1.axhline(y=over_50['thalach'].mean(), linestyle='dashdot')

## Adding title to the whole figure

fig.suptitle('Heart Disease Analysis', fontweight='bold', fontsize=20)
```
