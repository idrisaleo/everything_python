```python
# args, kwargs

def highest_even(l):
  evens = []
  for even in l:
    if even % 2 ==0:
      evens.append(even)
  return max(evens)
      
print(highest_even([8,34,56,78,90,98,76,54,33,2,4,4567,7,7,7890]))

## this is an example of *args and **kwargs. check the argument given below to help your brain

def whatever(*good, **bad):
  print(sum(good))
  total = 0
  for b in bad.values():
    total += b
    print(total)
  return total % sum(good)
    
print(whatever(1,2,3,4,5,6,5,6,7,8,7, ind=91, nig=234, gha=233,sau=966, chn=86))
```
